```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing data
cars = pd.read_csv('Toyota.csv', index_col=0, na_values=['??','????'])
#removing missing values
cars.dropna(axis=0,inplace= True)
cars.shape
```

```python
#creating scatter plot with regression line
sns.set(style='darkgrid')
sns.regplot(x=cars['Age'],y=cars['Price'])
```

```python
#creating scatter plot without regression line and * marker
sns.set(style='darkgrid')
sns.regplot(x=cars['Age'],y=cars['Price'], marker='*', fit_reg= False)
```

```python
#creating scatter plot with different colors
sns.lmplot(x='Age',y='Price', data=cars, fit_reg= False, hue= 'FuelType',legend=True, palette='Set1')
```

```python
#creating a histogram with kernel estimate
sns.displot(cars['Age'])
```

```python
#creating a histogram without kernel estimate
sns.displot(cars['Age'],kde=False)
```

```python
#creating a histogram without kernel estimate and with custom bin
sns.displot(cars['Age'],kde=False, bins=5)
```

```python
#creating bar plot
sns.countplot(x='FuelType',data=cars)
```

```python
#creating grouped bar plot
sns.countplot(x='FuelType',data= cars, hue ='Automatic')
```

```python
#creating box plot
sns.boxplot(y=cars['Price'])
```

```python
#creating a box plot to visualise more than one quantity
sns.boxplot(x=cars['FuelType'],y=cars['Price'])
```

```python
#grouped box plots
sns.boxplot(x=cars['FuelType'],y=cars['Price'], hue = 'Automatic',data=cars)
```

```python
#creating multiple plots in a single window
f,(ax_box,ax_hist)=plt.subplots(2,gridspec_kw={'height_ratios':(.15,.85)}) 
sns.boxplot(cars['Price'],ax=ax_box) 
sns.histplot(cars['Price'],ax=ax_hist,kde=False)
```

```python
#creating a pairwise plot
sns.pairplot(cars,kind='scatter',hue='FuelType')
plt.show()
```
