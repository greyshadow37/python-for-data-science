```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing data
cars = pd.read_csv('Toyota.csv', index_col=0, na_values = ['??','????'])
```

```python
#removing missing values
cars.dropna(axis = 0, inplace = True)
```

```python
#creating a scatter plot
plt.scatter(x=cars['Age'], y=cars['Price'], c='green')
#adding title
plt.title('Age vs Price')
#adding labels in x axis
plt.xlabel('Age in months')
#adding labels in y axis
plt.ylabel('Price in Euros')
#displaying the plot
plt.show()
```

```python
#creating a histogram
plt.hist(cars['KM'],
        color='blue',
        edgecolor='black',
        bins=5)
```

```python
#creating a bar plot
counts =[979,120,12]
fuelType=('Petrol','Diesel','CNG')
index=np.arange(len(fuelType))

plt.bar(index,counts,color=['crimson','violet','purple'])
plt.title('Frequency of Fuel Types')
plt.xlabel('Fuel Types')
#adding labels to each of the x elements
plt.xticks(index,fuelType,rotation=0)
plt.ylabel('Frequency')
plt.show()
```
